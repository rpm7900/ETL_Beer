```python
import pandas as pd
```

```python
#read in csv
breweries_df = pd.read_csv("CSV/breweries_and_brew_pubs_in_the_usa.csv")
```

```python
# #fix up zipcodes
for i, row in breweries_df.iterrows():
  if str(row['postalCode']) == 'nan':
    continue
  if str(row['postalCode']) == 'None':
    continue
  if len(str(row['postalCode'])) > 5:
    breweries_df['postalCode'].at[i] = str(row['postalCode'])[:5]
  if len(str(row['postalCode'])) < 5:
    while len(str(breweries_df['postalCode'].at[i])) < 5:
      breweries_df['postalCode'].at[i] = str("0{}".format(breweries_df['postalCode'].at[i]))
```

```python
#output into csv
export_csv = breweries_df.to_csv ('CSV/clean_breweries_and_brew_pubs_in_the_usa.csv', index = None, header=True)
```
